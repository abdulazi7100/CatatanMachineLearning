# Principal Component Analysis (PCA)

```python
import numpy as np
import pandas as pd

# Import the dataset
df = pd.read_csv('sample_data/california_housing_train.csv')
df
```

```python
# X = df[['total_rooms', 'households',	'median_income']]
X = df.drop(['total_bedrooms'], axis=1)
y = df['total_bedrooms']
X
```

# PCA SKLearn

```python
# Principal Component Analysis
from numpy import array
from sklearn.decomposition import PCA
# create the PCA instance
pca = PCA(8)
# fit on data
pca.fit(X)
```

```python
# access values and vectors
print(pca.explained_variance_)
```

```python
print(pca.components_)
```

```python
# transform data
X_pca = pca.transform(X)
print(X_pca)
```

# PCA from Scratch

```python
from numpy import array, mean, cov
from numpy.linalg import eig
```

```python
# calculate the mean of each column
M = mean(X, axis = 0)
M
```

```python
# center columns by subtracting column means
C = X - M
C
```

```python
# calculate covariance matrix of centered matrix
V = cov(C, rowvar = False)
V
```

```python
# eigendecomposition of covariance matrix
eigval, eigvec = eig(V)
```

```python
eigval
```

```python
eigvec
```

```python
#sort the eigenvalues in descending order
sorted_index = np.argsort(eigval)[::-1]
sorted_index
```

```python
sorted_eigenvalue = eigval[sorted_index]
sorted_eigenvalue
```

```python
#similarly sort the eigenvectors
sorted_eigenvectors = eigvec[:,sorted_index]
sorted_eigenvectors
```

```python
# select the first n eigenvectors, n is desired dimension
# of our final reduced data.

n_components = 2 #you can select any number of components.
eigenvector_subset = sorted_eigenvectors[:,0:n_components]
eigenvector_subset
```

```python
#Transform the data
X_reduced = np.dot(eigenvector_subset.transpose(),C.transpose()).transpose()
X_reduced
```
